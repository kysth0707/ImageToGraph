```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 흑백 이미지 읽기
img = cv2.imread('./testimgs/3.jpg', cv2.IMREAD_GRAYSCALE)

# 외곽선 검출
# RETR_EXTERNAL: 외곽선만 검출, CHAIN_APPROX_SIMPLE: 간소화된 경계점 표현
contours, hierarchy = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 빈 이미지 생성 (원본 이미지와 같은 크기)
outline_img = np.zeros_like(img)

# 외곽선 그리기 (1픽셀 두께)
cv2.drawContours(outline_img, contours, -1, (255), 10)  # -1: 모든 외곽선 그리기, 255: 흰색

# 결과 확인
plt.imshow(outline_img, cmap='gray')
```
