```python
import matplotlib.pyplot as plt
```

```python
import random
```

```python
dots = [
	(x, random.random()*10) for x in range(5)
]
```

```python
dotX = []
dotY = []
for x, y in dots:
	dotX.append(x)
	dotY.append(y)

plt.plot(dotX, dotY)
```

라그랑주 보간법

```python
list(set(dots) - set([(0, 5)]))
```

```python
dotSets = set(dots)

def g(x):
	sumOfValue = 0
	for dot in dots:
		elseDots = list(dotSets - set([dot]))
		# print(dot, elseDots)
		# print(i, dotX, dotY)
		Bunja = 1
		for elses in elseDots:
			Bunja *= x - elses[0]
		
		Bunmo = 1
		for elses in elseDots:
			Bunmo *= dot[0] - elses[0]
		sumOfValue += Bunja / Bunmo * dot[1]

	return sumOfValue
```

```python
import numpy as np
```

```python
dotX = []
dotY = []
for x, y in dots:
	dotX.append(x)
	dotY.append(y)

plt.plot(dotX, dotY)

t = list(np.arange(dotX[0], dotX[-1], 0.1))
y = [g(x) for x in t]
plt.plot(t, y)
```
